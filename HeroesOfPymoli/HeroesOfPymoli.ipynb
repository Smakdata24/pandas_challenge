```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
purchase_data = "Resources/purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df
```

```python
total_players = len(purchase_data_df["SN"].value_counts())

total_count = pd.DataFrame({"Total Players":[total_players]})
total_count                        
```

```python
total_unique = len(purchase_data_df["Item Name"].unique())
avg_price = purchase_data_df["Price"].mean()
total_purchase = purchase_data_df["Price"].count()
total_price = purchase_data_df["Price"].sum()


purchasing_analysis = pd.DataFrame({
    "Number of Unique Items":[total_unique],
    "Average Purchase Price":[avg_price],
    "Total Number of Purchases":[total_purchase],
    "Total Revenue":[total_price]
})
purchasing_analysis



purchasing_analysis.style.format({"Average Purchase Price":"${:,.2f}",
                         "Total Revenue": '${:,.2f}'})
```

```python
gender_analysis = purchase_data_df.groupby("Gender")

gender_total = gender_analysis.nunique()["SN"]
gender_total_f = gender_total.astype(float).map("{:,.2f}".format)

percent_total = gender_total / (576) * 100
percent_total_f = percent_total.astype(float).map("{:,.2f}%".format)

total_players = pd.DataFrame({"Total Count": gender_total, 
                              "Percentage of Players": percent_total
                             })


total_players_sorted = total_players.sort_values(["Total Count", "Percentage of Players"], ascending = False).style.format({
    "total_players_sorted":"{:,.2f}",
    "Percentage of Players": "{:,.2f}%"})

total_players_sorted

```

```python
purchase_count = gender_analysis["Purchase ID"].count()
avg_purchase = gender_analysis["Price"].mean()
total_purchase = gender_analysis["Price"].sum()
avg_total = total_purchase / gender_total
avg_total

purchasing_analysis_df = pd.DataFrame({"Purchase Count": purchase_count,
                                    "Average Purchase Price": avg_purchase,
                                    "Total Purchase Value": total_purchase,
                                    "Average Total Purchase per Person": avg_total
                                   })
purchasing_analysis_df

purchasing_analysis_df["Purchase Count"] = purchasing_analysis_df["Purchase Count"].astype(float).map(
    "${:,.2f}".format)
purchasing_analysis_df["Average Purchase Price"] = purchasing_analysis_df["Average Purchase Price"].astype(float).map(
    "${:,.2f}".format)
purchasing_analysis_df["Total Purchase Value"] = purchasing_analysis_df["Total Purchase Value"].astype(float).map(
    "${:,.2f}".format)
purchasing_analysis_df["Average Total Purchase per Person"] = purchasing_analysis_df["Average Total Purchase per Person"].astype(float).map(
    "${:,.2f}".format)

purchasing_analysis_df
```

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Demographics"] = pd.cut(purchase_data_df["Age"],
                                        age_bins, labels=group_age, include_lowest=True)

age_group = purchase_data_df.groupby("Age Demographics")
total_age_group = age_group["SN"].nunique()
Percent_age_group = total_age_group / (576) * 100
age_group

age_demo_df = pd.DataFrame({"Total Count": total_age_group,
                            "Percentage of Players": Percent_age_group
                                                      })

age_demo_df


age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

age_demo_df
```

```python
age_purchase_count = age_group["Purchase ID"].count()
age_purchase_count

avg_purchase_price = age_group["Price"].mean()
avg_purchase_price

total_purchase_value = age_group["Price"].sum()
total_purchase_value

avg_total_per_person = total_purchase_value / total_age_group
avg_total_per_person


purchasing_analysis_df = pd.DataFrame({"Purchase Count": age_purchase_count,
                                      "Average Purchase Price":avg_purchase_price,
                                      "Total Purchase Value": total_purchase_value,
                                      "Avg Total Purchase per Person": avg_total_per_person
                                      })
purchasing_analysis_df

purchasing_analysis_df["Average Purchase Price"] = purchasing_analysis_df["Average Purchase Price"].astype(float).map(
    "${:,.2f}".format)

purchasing_analysis_df["Total Purchase Value"] = purchasing_analysis_df["Total Purchase Value"].astype(float).map(
    "${:,.2f}".format)

purchasing_analysis_df["Avg Total Purchase per Person"] = purchasing_analysis_df["Avg Total Purchase per Person"].astype(
    float).map("${:,.2f}".format)

purchasing_analysis_df
```
